//imports and data loading 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/netflix1.csv")
df.head()
df.info()
df.describe()
```

//data cleaning//

```python
print(df.duplicated().sum())
```

//conversion of date to datetime format

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['release_year'].head()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'])
df['release_year'].head()
```

//exploratory data analysis

```python
df['type'].value_counts()
```

```python
plt.figure(figsize=(16,6))
df['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title("Content Release Year Distribution")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='barh')
```

```python
df['rating'].value_counts().plot(kind='line')
```

//content added to netflix over time//

```python
df['date_added'].dt.year.value_counts().sort_index().plot(kind='line')
```

//visualization

```python
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows")
plt.show()
```
